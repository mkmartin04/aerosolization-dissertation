## OC-CCI mapped analysis for chlorophyll

```python
%matplotlib inline
# Plotting
import matplotlib.pyplot as plt
from matplotlib import gridspec
import matplotlib.dates as dates
from matplotlib.dates import DateFormatter
# Statistics
import pymannkendall as mk
import numpy as np
import pandas as pd
# NetCDF
import netCDF4 as nc
# Import and ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Accessing the product
in_file = 'WCS753932556675586793.nc'
nc_data = nc.Dataset(in_file)
nc_data
```

```python
# Extract variables time, lat, lon, cdom_412
time= nc_data.variables['time']
lat = np.array(nc_data.variables['lat'] )
lon = np.array(nc_data.variables['lon'] )
chl = np.array(nc_data.variables['chlor_a'])
print(chl.shape)
```

```python
# Get your dates into a user-friendly format: dtime is your new date variable
caldr = time.getncattr('calendar')
units = time.getncattr('units')
dtime = nc.num2date(time[:], units, calendar = caldr, only_use_cftime_datetimes = False)
dtime
```

## Point measurement: specify pixel 

```python
# Central pixel inside the bay
LAT_p = 36.38
LON_p = 120.76
# change above to match the area you want
lat_ind = np.argmin(np.abs(lat - LAT_p))   
lon_ind = np.argmin(np.abs(lon - LON_p))
print(lat_ind,lon_ind)
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create figure & specify dimensions:
fig1 = plt.figure(figsize = (11, 11))

# Define data & output projections:
data_projection= ccrs.PlateCarree()
out_projection = ccrs.PlateCarree()

# Add subplot with specified map projection
ax = plt.axes(projection = out_projection)

# Add gridlines and LAT / LON labels to left and bottom of your map
gridlines = ax.gridlines(crs = data_projection, draw_labels = True, 
                         linewidth = 1, color ='white', alpha = 0.5, linestyle = ':')
gridlines.top_labels  = False
gridlines.right_labels= False
gridlines.xlabel_style= {'size': 12, 'color': 'dimgray'}
gridlines.ylabel_style= {'size': 12, 'color': 'dimgray'}

# Add land and coastline with facecolor and coastline filled in -- black
land = cfeature.NaturalEarthFeature('physical', 'land', '10m',
                                    edgecolor = 'black', facecolor = 'black')
ax.add_feature(land)

# Plot the SST data using pcolormesh with the 'viridis' color map
im = ax.pcolormesh(lon, lat, chl[7], cmap='viridis', transform = data_projection)
# Set SST limits (min, max)
#im.set_clim(21, 28)

# Add contour lines for emphasis
contours = ax.contour(lon, lat, chl[7], levels = 10, 
                      transform = data_projection, cmap = 'Greys', linewidths = 1)
ax.clabel(contours, fontsize = 20, inline = True)

# Add single point overlay
ax.scatter(LON_p, LAT_p, transform=ccrs.PlateCarree(), marker='*', c= 'red', s=400)

# Add a colorbar with label
cbar = plt.colorbar(im, orientation='horizontal', fraction = 0.05, pad = 0.04)
cbar.ax.tick_params(labelsize=11)  # Set the fontsize for the colorbar
plt.title('Chlorophyll', fontsize = 11, fontweight = 'bold')

# Display the final plot
plt.show()
```
